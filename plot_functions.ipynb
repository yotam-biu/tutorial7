```python
import numpy as np # For generate list of numbers
import math # For sin function
import matplotlib.pyplot as plt # For plot
```

# $\frac{\sin{x}}{x}$

```python
xs_with_zero = np.linspace(-10,10,100)
xs = []
for x in xs_with_zero:
    if x != 0:
        xs.append(x)

ys = []
for x in xs:
    ys.append(math.sin(x)/x)

plt.plot(xs,ys,marker='o')
plt.show()
```

# $ x \sin{\frac{1}{x}} $

```python
xs_with_zero = np.linspace(-0.1,0.1,1000)
xs = []
for x in xs_with_zero:
    if x != 0:
        xs.append(x)


ys = []
for x in xs:
    ys.append(x * math.sin(1/x))

plt.plot(xs,ys,marker='o')
plt.show()
```

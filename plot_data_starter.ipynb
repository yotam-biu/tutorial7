```python
import random
from matplotlib import pyplot as plt
```

# Zip

```python
xs = ['A', 'B', 'C', 'D']
ys = [1,10,100,1000]

for x, y in zip(xs, ys):
    print(x, y)
```

# Histogram

```python


def generate_random_numbers(n):
  data = []
  for _ in range(n):
    data.append(random.random())

  return data
```

```python
data = generate_random_numbers(100)
plt.hist(data, bins = 10, edgecolor = 'black')
plt.title('100 random numbers')
plt.show()
```

```python


fig, axs = plt.subplots(1,3,figsize=(20,5))


data = generate_random_numbers(100)
axs[0].hist(data, bins=10, edgecolor='black')
axs[0].set_title('100 random numbers')

data = generate_random_numbers(1000)
axs[1].hist(data, bins=10, edgecolor='black')
axs[1].set_title('1000 random numbers')

data = generate_random_numbers(10000)
axs[2].hist(data, bins=10, edgecolor='black')
axs[2].set_title('10000 random numbers')


plt.show()
```

```python

n_items = [10**1, 10**2, 10**3, 10**4, 10**5]


fig, axs = plt.subplots(1, len(n_items),figsize=(20,3))


for ax, n in zip(axs, n_items):
  data = generate_random_numbers(n)
  ax.hist(data, bins=10, edgecolor='black')
  ax.set_title(str(n) + 'random numbers')

plt.show()
```

# Correlation

```python
x = list(range(10))

y_random = []
for i in x:
    y_random.append(random.random())

y_corr = []
for i in x:
    y_corr.append(-i + random.random())


fig, axs = plt.subplots(1, 2, figsize = (10,5))
axs[0].scatter(x, y_random)
axs[0].set_title('Random')
axs[0].set_xlabel('x', fontsize = 20)
axs[0].set_ylabel('y', fontsize = 20)

axs[1].scatter(x, y_corr)
axs[1].set_title('Correlated Data')
axs[1].set_xlabel('x', fontsize = 20)
axs[1].set_ylabel('y', fontsize = 20)
plt.show()
```
